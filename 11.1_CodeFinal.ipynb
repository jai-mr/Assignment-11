```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# 2 x step_size = 1 cycle
num_cycles = 6
step_size = 20
```

```python
lr_min = 0.001
lr_max = 0.010
```

```python
total_iters = step_size * 2 * num_cycles
```

```python
def get_lr(iter):
    cycle = np.floor(1 + iter / (2 * step_size))
    x = np.abs( iter/step_size - 2 * cycle + 1 )
    lr_t = lr_min + (lr_max - lr_min) * (1 - x)
    return lr_t
```

```python
x = np.linspace(0, total_iters, 1000);
y = np.array([get_lr(iter) for iter in x])
```

```python
#plt.style.use("dark_background")
plt.style.use("grayscale")

plt.figure(figsize=(12,6))
plt.plot(x, y)
plt.xlabel('step_t')
plt.ylabel('lr_t')
plt.show()
```
